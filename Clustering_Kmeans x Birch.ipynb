```python
# 1. Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans, Birch
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
from scipy.sparse import hstack
```

```python
# 2. Load Dataset
df = pd.read_csv('/content/pricerunner_aggregate.csv', sep=';')
df.head()
```

```python
# 3. Data Cleaning & Persiapan
df.columns = df.columns.str.strip()
df['Product Title'] = df['Product Title'].fillna('')
```

```python
# 4. Feature Engineering
le_merchant = LabelEncoder()
le_category = LabelEncoder()

df['Merchant_ID'] = le_merchant.fit_transform(df['Merchant ID'])
df['Category_ID'] = le_category.fit_transform(df['Category Label'])

tfidf = TfidfVectorizer(max_features=300)
X_text = tfidf.fit_transform(df['Product Title'])

X_merchant = df[['Merchant_ID']].values
X_category = df[['Category_ID']].values
```

```python
# 5. Gabungkan Semua Fitur
from scipy import sparse
X_combined = hstack([X_text, X_merchant, X_category])
```

```python
# 6. Evaluasi KMeans (k = 2 sampai 15)
silhouette_scores = []
inertia_scores = []
k_range = range(2, 16)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_combined)
    inertia_scores.append(kmeans.inertia_)
    sil_score = silhouette_score(X_combined, labels)
    silhouette_scores.append(sil_score)

# Tentukan best_k (nilai k dengan Silhouette Score tertinggi)
best_k = k_range[np.argmax(silhouette_scores)]
print(f"Best k: {best_k}")
```

```python
# 7. Visualisasi Elbow dan Silhouette Score
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_range, inertia_scores, marker='o')
plt.title("Elbow Method")
plt.xlabel("k")
plt.ylabel("Inertia")

plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o', color='green')
plt.title("Silhouette Score")
plt.xlabel("k")
plt.ylabel("Score")

plt.tight_layout()
plt.show()
```

```python
# 8. Fit Final KMeans dengan best_k
kmeans_final = KMeans(n_clusters=best_k, random_state=42)
kmeans_labels = kmeans_final.fit_predict(X_combined)
df['Cluster_KMeans'] = kmeans_labels

# Evaluasi KMeans
sil_kmeans = silhouette_score(X_combined, kmeans_labels)
dbi_kmeans = davies_bouldin_score(X_combined.toarray(), kmeans_labels)

print(f"Silhouette Score (KMeans): {sil_kmeans}")
print(f"Davies-Bouldin Index (KMeans): {dbi_kmeans}")
```

```python
# 9. Visualisasi Clustering KMeans (PCA)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_combined.toarray())

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.5)
plt.title(f"KMeans Clustering Visualization (k={best_k})")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label="Cluster")
plt.show()
```

```python
# 10. Clustering dengan Birch
birch = Birch(n_clusters=best_k)
birch_labels = birch.fit_predict(X_combined)
df['Cluster_Birch'] = birch_labels

# Evaluasi Birch
sil_birch = silhouette_score(X_combined, birch_labels)
dbi_birch = davies_bouldin_score(X_combined.toarray(), birch_labels)

print(f"Silhouette Score (Birch): {sil_birch}")
print(f"Davies-Bouldin Index (Birch): {dbi_birch}")
```

```python
# 11. Visualisasi Clustering Birch (PCA)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=birch_labels, cmap='plasma', alpha=0.5)
plt.title(f"Birch Clustering Visualization (k={best_k})")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label="Cluster")
plt.show()
```

```python
# 12. Simpan Hasil ke CSV
df.to_csv('/content/hasil_clustering_produk.csv', index=False, sep=';')
print("✅ Hasil clustering disimpan ke /content/hasil_clustering_produk.csv")
```
